```python
payment = df.loc[:, 'pharmacy':'rent_car']
act = df.loc[:, 'act_pharmacy':'act_rent_car']
cashback = df.loc[:, 'back_pharmacy':'back_rent_car']
for i in range(17):
    ids = df.loc[(act.iloc[:, i] == 1) & (payment.iloc[:, i] <= cashback.iloc[:, i])]['client_id'].to_list()
    df = df.loc[~df['client_id'].isin(ids)]
```

```python
zeroes = df.groupby('client_id')['total'].sum() == 0
df = df.loc[~df['client_id'].isin(zeroes[zeroes].index)]
```

```python
print(len(df))
print(59826 - len(df))
```

```python
nums = [len(df), 59826 - len(df)]
labels = ['Осталось', 'Удалено']
plt.style.use("dark_background")
cols = ['gold', 'white']
_, _, autotexts = plt.pie(nums, labels=labels, colors=cols, autopct='%1.1f%%')
for autotext in autotexts:
    autotext.set_color('black')
plt.savefig("graphs/deleted.svg")
plt.show()
```

```python
num_of_activations = df.loc[:, 'act_pharmacy':'act_rent_car']
sum_of_act = num_of_activations.sum(axis=1)
labels = [0, 1, 2, 3, 4]
data = []
for i in range(5):
    data.append(sum_of_act.loc[sum_of_act == i].count())
    
plt.style.use("dark_background")
plt.tight_layout()

ser = pd.Series(data, index=labels)
ser = ser.to_frame()

barSet = sns.barplot(x=ser.index, y=ser[0], data=ser, errorbar=None)
barSet.set(xlabel='кол-во кэшбэков', ylabel='кол-во строк')
barSet.bar_label(barSet.containers[0])
plt.savefig("graphs/кэшбэк.svg")
plt.show()
```
