```python
df
```

```python
payment = df.loc[:, 'pharmacy':'rent_car']
act = df.loc[:, 'act_pharmacy':'act_rent_car']
cashback = df.loc[:, 'back_pharmacy':'back_rent_car']
for i in range(17):
    ids = df.loc[(act.iloc[:, i] == 1) & (payment.iloc[:, i] <= cashback.iloc[:, i])]['client_id'].to_list()
    df = df.loc[~df['client_id'].isin(ids)]
```

```python
nums = [len(df), 59826 - len(df)]
labels = ['Осталось', 'Удалено']
plt.style.use("dark_background")
cols = ['gold', 'white']
_, _, autotexts = plt.pie(nums, labels=labels, colors=cols, autopct='%1.1f%%')
for autotext in autotexts:
    autotext.set_color('black')
plt.savefig("deleted.svg")
plt.show()
```
